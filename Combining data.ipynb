```python
import pandas as pd
import numpy as np
import os
```

собираю данные с разных страниц в один файл

```python
path = 'page data/'

all_csv_data = []

for filename in os.listdir(path):
    if filename.endswith('.csv'):
        df = pd.read_csv(os.path.join(path, filename), sep=';', encoding='cp1251')
        all_csv_data.append(df)

cian = pd.concat(all_csv_data, axis=0, ignore_index=True)
```

```python
cian.drop_duplicates()
```

```python
cian.link.nunique()
```

```python
cian.drop(['phone', 'year_of_construction', 'kitchen_meters', 'living_meters'], axis=1, inplace=True)
```

```python
cian.drop_duplicates(inplace=True)
```

```python
cian.to_csv('cian_to_true.csv', sep=';', index = False)
```
